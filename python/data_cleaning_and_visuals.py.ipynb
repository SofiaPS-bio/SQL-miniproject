```python
def standardize_columns(df, year):
    """
    Standardize column names across different years of happiness data.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The dataframe containing happiness data
    year : int
        The year of the data to be added as a column
    
    Returns:
    --------
    pandas.DataFrame
        Dataframe with standardized column names and added year column
    """
    # Dictionary mapping various column names to standardized names
    # This handles inconsistencies in column naming across different years' datasets
    rename_map = {
        'Country': 'Country',
        'Country or region': 'Country',
        'Happiness Rank': 'Rank',
        'Happiness.Rank': 'Rank',
        'Overall rank': 'Rank',
        'Happiness Score': 'Score',
        'Happiness.Score': 'Score',
        'Score': 'Score',
        'Economy (GDP per Capita)': 'GDP_per_capita',
        'Economy..GDP.per.Capita.': 'GDP_per_capita',
        'GDP per capita': 'GDP_per_capita',
        'Family': 'Social_support',
        'Social support': 'Social_support',
        'Health (Life Expectancy)': 'Healthy_life_expectancy',
        'Health..Life.Expectancy.': 'Healthy_life_expectancy',
        'Healthy life expectancy': 'Healthy_life_expectancy',
        'Freedom': 'Freedom',
        'Freedom to make life choices': 'Freedom',
        'Generosity': 'Generosity',
        'Trust (Government Corruption)': 'Trust',
        'Trust..Government.Corruption.': 'Trust',
        'Perceptions of corruption': 'Trust'
    }

    # Rename columns using the mapping dictionary
    df = df.rename(columns=rename_map)
    # Add a year column to the dataframe
    df['Year'] = year
    return df
```

```python
df2015 = standardize_columns(df2015, 2015)
df2016 = standardize_columns(df2016, 2016)
df2017 = standardize_columns(df2017, 2017)
df2018 = standardize_columns(df2018, 2018)
df2019 = standardize_columns(df2019, 2019)
```

```python
expected_cols = ['Country', 'Rank', 'Score', 'GDP_per_capita', 'Social_support',
                 'Healthy_life_expectancy', 'Freedom', 'Generosity', 'Trust', 'Year']

for df, year in zip([df2015, df2016, df2017, df2018, df2019], [2015, 2016, 2017, 2018, 2019]):
    print(f"{year}: {set(df.columns) & set(expected_cols)}")
```

```python
# 1. List to accumulate the harmonized DataFrames
clean_dataframes_list = []

# 2. Define the standard column list (Standardized for SQL)
common_columns = [
    'Country',
    'Score',
    'GDP_per_capita',
    'Social_support',
    'Healthy_life_expectancy',
    'Freedom',
    'Generosity',
    'Trust',
    'Year'
]

# 3. Mapping of inconsistent columns (Mandatory for harmonization)
# This mapping ensures columns with different names are aligned
column_map = {
    # Older / inconsistent names : Unique Final Name
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP_per_capita',
    'Family': 'Social_support',
    'Health (Life Expectancy)': 'Healthy_life_expectancy',
    'Trust (Government Corruption)': 'Trust',
    'Overall rank': 'Rank', # We exclude 'Rank' from final columns, but it's good to map
    'Country or region': 'Country',
    'Perceptions of corruption': 'Trust',
    'Social support': 'Social_support', # For 2018/2019
    'Healthy life expectancy': 'Healthy_life_expectancy', # For 2018/2019
    'Freedom to make life choices': 'Freedom' # For 2018/2019
}

# 4. Create a list of the dataframes and their corresponding years
# NOTE: Assumes DFs (df2015 to df2019) were loaded into memory before running this block.
dataframes_and_years = [
    (df2015, 2015),
    (df2016, 2016),
    (df2017, 2017),
    (df2018, 2018),
    (df2019, 2019)
]

# 5. Iterate, harmonize, and accumulate
for df, year in dataframes_and_years:
    
    print(f"Processing data for {year}...")
    
    # Add the 'Year' column first (Ensuring it is present and correct)
    df['Year'] = year
    
    # Harmonize: Rename all problematic columns at once
    df = df.rename(columns=column_map)
    
    # Select only the final standard columns (This also removes extra columns like 'Rank' and 'Residual')
    try:
        df_harmonized = df[common_columns]
        
        # Add the clean, harmonized DataFrame to the list
        clean_dataframes_list.append(df_harmonized)
        
    except KeyError as e:
        # If a required column is missing from the mapping
        print(f"Required column was not found: {e}. Check the mapping for {year}.")
        continue # Skip to the next DataFrame
        
# 6. Combine all accumulated DataFrames
df = pd.concat(clean_dataframes_list, ignore_index=True)

# 7. Export the cleaned DataFrame to a CSV file
df.to_csv('happiness_score.csv', index=False)
```

# Data Enrichment: Integration of Economic and Innovation Metrics
Objective: To expand the analytical scope of the existing master World Happiness dataset.

Procedure: The newly consolidated happiness data will be augmented by merging it with three external, supplementary datasets:

GDP (Gross Domestic Product)

GDP Per-Capita

Innovation Score

Outcome: This merge will yield a single, comprehensive dataset, enabling statistically rigorous analysis and the creation of clear visualizations that explore the relationship between national happiness levels, economic performance, and innovation capacity.

```python
df_h = pd.read_csv('happiness_scores.csv')
```

```python

# ----------------------------------------------------------------------
# PHASE 1: COUNTRY MAPPING AND STANDARDIZATION
# Goal: Add the 'Code' column (3-letter country code) to df_h, which only has 'Country' name.
# We use df_gpc as the source for the Country and Code mapping.
# ----------------------------------------------------------------------

# 1. Create the Country <-> Code Mapping Table
country_code_map = df_gpc[['Country', 'Code']].drop_duplicates()

# 2. Add the 'Code' column to the Happiness Data (df_h)
# We create a new DataFrame 'df_h_coded'
df_h_coded = pd.merge(
    df_h,
    country_code_map,
    on='Country',
    how='left'  # Keep all happiness records.
)

# CRITICAL CHECK: Identify unmatched countries (names that differ)
unmatched_count = df_h_coded['Code'].isna().sum()

if unmatched_count > 0:
    print(f"ATTENTION: {unmatched_count} happiness records could not be matched by 'Country' name.")
    print("These will have NaN in the 'Code' column and should be manually reconciled or dropped later.")
    
# Drop the original GDP_per_capita column from df_h, as we will use the one from df_gpc for consistency.
df_h_coded = df_h_coded.drop(columns=['GDP_per_capita'], errors='ignore')


# ----------------------------------------------------------------------
# PHASE 2: WIDE-TO-LONG CONVERSION (MELT)
# Goal: Convert annual data (df_g, df_gpc, df_i) from wide format (years as columns)
#       to long format ('Year' and 'Value' columns).
# ----------------------------------------------------------------------

# 3. Convert GDP Data (df_g) to Long Format
gdp_long = pd.melt(
    df_g,
    id_vars=['Country', 'Code'],
    var_name='Year',
    value_name='GDP_Total'
).dropna(subset=['GDP_Total'])
gdp_long['Year'] = gdp_long['Year'].astype(int)

# 4. Convert GDP Per Capita Data (df_gpc) to Long Format
gdp_per_capita_long = pd.melt(
    df_gpc,
    id_vars=['Country', 'Code'],
    var_name='Year',
    value_name='GDP_per_capita'
).dropna(subset=['GDP_per_capita'])
gdp_per_capita_long['Year'] = gdp_per_capita_long['Year'].astype(int)

# 5. Convert Innovation Data (df_i) to Long Format

# Melt Scores
innovation_scores_long = pd.melt(
    df_i,
    id_vars=['Code'],
    value_vars=[col for col in df_i.columns if 'Score' in col],
    var_name='Score_Year',
    value_name='Innovation_Score'
)
innovation_scores_long['Year'] = innovation_scores_long['Score_Year'].str.replace('Score ', '').astype(int)

# Melt Ranks
innovation_ranks_long = pd.melt(
    df_i,
    id_vars=['Code'],
    value_vars=[col for col in df_i.columns if 'Rank' in col],
    var_name='Rank_Year',
    value_name='Innovation_Rank'
)
innovation_ranks_long['Year'] = innovation_ranks_long['Rank_Year'].str.replace('Rank ', '').astype(int)

# Merge Innovation Scores and Ranks back together
innovation_long = pd.merge(
    innovation_scores_long[['Code', 'Year', 'Innovation_Score']],
    innovation_ranks_long[['Code', 'Year', 'Innovation_Rank']],
    on=['Code', 'Year'],
    how='outer'
)


# ----------------------------------------------------------------------
# PHASE 3: FINAL MERGE
# Goal: Combine all processed dataframes using the common keys 'Code' and 'Year'.
# ----------------------------------------------------------------------

# 6. Merge GDP Total and GDP Per Capita
economic_df = pd.merge(
    gdp_long[['Code', 'Year', 'GDP_Total']],
    gdp_per_capita_long[['Code', 'Year', 'GDP_per_capita']],
    on=['Code', 'Year'],
    how='outer'
)

# 7. Merge Economic Data with Innovation Data
economic_innovation_df = pd.merge(
    economic_df,
    innovation_long,
    on=['Code', 'Year'],
    how='outer' # Use outer to keep all economic/innovation data
)

# 8. Final Merge: Happiness Data with All Others
final_df = pd.merge(
    df_h_coded,
    economic_innovation_df,
    on=['Code', 'Year'],
    how='left' # Keep all happiness data and link the economic/innovation data
)

# Clean up final dataframe (remove rows where Code was NaN due to unmatched country names)
final_df = final_df.dropna(subset=['Code'])

print("\n--- Final DataFrame Information ---")
print("Data Preparation Complete. Ready for analysis.")
print(f"Total rows in final dataset: {len(final_df)}")
print("\nFinal DataFrame Columns:")
print(final_df.columns.tolist())
```

```python
# List of columns for which we want to create a ranking
ranking_columns = [
    'Score',
    'Social_support',
    'Healthy_life_expectancy',
    'Freedom',
    'Generosity',
    'Trust',
    'GDP_Total',
    'GDP_per_capita'
]

print("Calculating annual ranks for selected metrics...")

for col in ranking_columns:
    # Define the name of the new rank column
    new_rank_col = f'Rank_{col}'
    
    # Calculate the rank per year:
    # 1. GroupBy('Year'): Ensures ranking is done only among countries in the same year.
    # 2. rank(): Computes the rank.
    # 3. ascending=False: Highest value (e.g., highest Score) receives Rank 1 (best rank).
    # 4. method='min': In case of ties, all tied values receive the best rank possible (e.g., Rank 2).
    
    final_df[new_rank_col] = final_df.groupby('Year')[col].rank(
        method='min',
        ascending=False
    )
    
    # Convert the rank to integer type, excluding NaNs, as ranks should be whole numbers.
    # Using 'Int64' (capital I) allows for nullable integers (keeping NaNs where data is missing).
    final_df[new_rank_col] = final_df[new_rank_col].astype('Int64')

print("--- Ranking Calculation Complete ---")
print(f"New rank columns successfully added to 'final_df'.")
```

```python
final_df.head()
```

```python
# Create a list of the base columns (Identification + Year) in the desired starting order
base_columns = ['Country', 'Code', 'Year']

# List of all measurable metrics (those that have both a raw value and a rank)
metric_columns = [
    'Score', 'Social_support', 'Healthy_life_expectancy', 'Freedom',
    'Generosity', 'Trust', 'GDP_Total', 'GDP_per_capita',
    'Innovation_Score', 'Innovation_Rank'
    # Note: Innovation_Rank is already a rank, we will keep it next to Innovation_Score.
]

# Create the new column order list
new_column_order = []

# 1. Add the base columns
new_column_order.extend(base_columns)

# 2. Add metrics and their corresponding ranks
for col in metric_columns:
    # Check if the column is one of the metrics we created a rank for
    if col in ['Score', 'Social_support', 'Healthy_life_expectancy', 'Freedom', 'Generosity', 'Trust', 'GDP_Total', 'GDP_per_capita']:
        new_column_order.append(col)
        # Add the rank column right after the metric column
        rank_col = f'Rank_{col}'
        if rank_col in final_df.columns:
            new_column_order.append(rank_col)
    else:
        # Add the remaining metric columns (like the Innovation pair)
        new_column_order.append(col)


# Final adjustment: Ensure all columns are present (e.g., if any column was missed)
# This handles cases where a column might have been missed in the metric_columns list.
for col in final_df.columns:
    if col not in new_column_order:
        new_column_order.append(col)


# Reindex the DataFrame using the new order
final_df = final_df.reindex(columns=new_column_order)

print("--- Data Reordering Complete ---")
print(final_df.columns.tolist())
```

```python
# Dictionary defining the old and new column names
rename_map = {
    'Score': 'Happiness_Score',
    'Rank_Score': 'Rank_Happiness_Score'
}

# Use the rename method on the final_df
final_df = final_df.rename(columns=rename_map)


# Print the new column list to confirm the change and show the new order
print(final_df.columns.tolist()[:5])
```

```python
# Create a list of the key columns where missing data impacts analysis
key_metrics = ['GDP_per_capita', 'Innovation_Score']

# Group by Country and count the number of missing values (NaN) for the key metrics
# A value of 5 means the country is missing data for that metric for ALL 5 years (2015-2019).
missing_data_summary = final_df.groupby('Country')[key_metrics].apply(lambda x: x.isnull().sum())

# Calculate the total number of missing data points per country
missing_data_summary['Total_Missing'] = missing_data_summary.sum(axis=1)

# Sort to find the worst offenders (countries we failed to fully map or where data is scarce)
worst_offenders = missing_data_summary.sort_values(by='Total_Missing', ascending=False).head(20)

print("--- Pases com a Maior Perda/Lacunas de Dados (2015-2019) ---")
print("Isto indica quais pases esto no final_df, mas no contribuem para a anlise de Correlao:")
print("\n" + worst_offenders.to_string())

print("\n(Um valor de 10 significa que o pas no tem PIB E Inovao em 5 anos.)")
```

```python
# Assuming 'missing_data_summary' DataFrame from the previous step is still available.

# 1. Identify countries to drop (where total missing values in key metrics >= 5)
countries_to_drop = missing_data_summary[missing_data_summary['Total_Missing'] >= 5].index.tolist()

# 2. Filter the final_df to keep only the good records
rows_before = len(final_df)
final_df = final_df[~final_df['Country'].isin(countries_to_drop)]
rows_after = len(final_df)

# 3. Print confirmation
print("--- Final Data Pruning Complete ---")
print(f"Removed {len(countries_to_drop)} countries due to extensive missing data in key metrics.")
print(f"Total rows removed: {rows_before - rows_after}")
print(f"New total rows in final_df: {rows_after}")

# 4. Final check on the number of countries per year
print("\n--- New Country Counts per Year ---")
print(final_df.groupby('Year')['Country'].nunique())
```

```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Funo para Plotar os Coeficientes ---
def plot_coefficients_comparison(model, title, color_list):
    """
    Ajusta o modelo, extrai os coeficientes e plota-os em um grfico de barras horizontal.
    """
    # 1. Extrair os coeficientes e remover o intercepto
    coefficients = model.params.drop('const')
    
    # 2. Ordenar do mais importante para o menos importante
    coefficients = coefficients.sort_values(ascending=False)
    
    # 3. Mapeamento de nomes de colunas para nomes amigveis
    name_map = {
        'Healthy_life_expectancy': 'Healthy Life Expectancy', 
        'Social_support': 'Social Support', 
        'Trust': 'Trust (Absence of Corruption)', 
        'Freedom': 'Freedom to Make Life Choices'
    }
    coefficients.index = [name_map.get(idx, idx) for idx in coefficients.index]

    # 4. Criao do Grfico
    plt.figure(figsize=(10, 6))
    # Usar 'barh' (barras horizontais) para melhor visualizao dos nomes
    coefficients.plot(kind='barh', color=color_list[:len(coefficients)])
    
    plt.title(title, fontsize=14)
    plt.xlabel('Coefficient Value (Impact on Happiness Score)', fontsize=10)
    plt.ylabel('Factors')
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.gca().invert_yaxis() # Coloca o maior coeficiente em cima
    plt.show()

# --- DEFINIO DOS MODELOS (Ajustar novamente para garantir) ---

# Re-executar o ajuste de cada modelo (para garantir que os objetos existam)

# Modelo 1: Trust
X1 = final_df_clean[['Healthy_life_expectancy', 'Social_support', 'Trust']]
Y = final_df_clean['Happiness_Score']
X1 = sm.add_constant(X1)
model_1 = sm.OLS(Y, X1).fit()

# Modelo 2: Freedom
X2 = final_df_clean[['Healthy_life_expectancy', 'Social_support', 'Freedom']]
X2 = sm.add_constant(X2)
model_2 = sm.OLS(Y, X2).fit()

# Modelo 3: Completo
X3 = final_df_clean[['Healthy_life_expectancy', 'Social_support', 'Trust', 'Freedom']]
X3 = sm.add_constant(X3)
model_3 = sm.OLS(Y, X3).fit()


# --- GERAO DOS GRFICOS ---
# Usamos cores ligeiramente diferentes para ajudar na distino

# Grfico 1: Modelo com Confiana
plot_coefficients_comparison(
    model_1, 
    'Model 1: Impact of Factors on Happiness Score (Trust)', 
    ['#4daf4a', '#ff7f00', '#377eb8']
)

# Grfico 2: Modelo com Liberdade
plot_coefficients_comparison(
    model_2, 
    'Model 2: Impact of Factors on Happiness Score (Freedom)', 
    ['#4daf4a', '#e41a1c', '#ff7f00']
)

# Grfico 3: Modelo Completo (Todos os fatores)
plot_coefficients_comparison(
    model_3, 
    'Model 3: Impact of Factors on Happiness Score', 
    ['#4daf4a', '#377eb8', '#e41a1c', '#ff7f00']
)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd

# --- 1. Ajustar Modelos de Regresso Simples ---

# Modelo Simples: Inovao vs. GDP_per_capita
X_gdp_const = sm.add_constant(final_df_clean['GDP_per_capita'])
model_gdp_innovation = sm.OLS(final_df_clean['Innovation_Score'], X_gdp_const).fit()
coef_gdp = model_gdp_innovation.params['GDP_per_capita']

# Modelo Simples: Inovao vs. Happiness_Score
X_happiness_const = sm.add_constant(final_df_clean['Happiness_Score'])
model_happiness_innovation = sm.OLS(final_df_clean['Innovation_Score'], X_happiness_const).fit()
coef_happiness = model_happiness_innovation.params['Happiness_Score']

# --- 2. Preparar Dados para Plotagem ---
df_impact = pd.DataFrame({
    'Factor': ['GDP per Capita', 'Happiness Score'],
    'Impact Coefficient on Innovation': [coef_gdp, coef_happiness]
})

# --- 3. Criar Grfico de Barras de Impacto ---
plt.figure(figsize=(9, 6))
sns.barplot(
    x='Impact Coefficient on Innovation', 
    y='Factor', 
    data=df_impact, 
    palette=['green', 'red'], # Cores personalizadas
    orient='h' # Barras horizontais
)
plt.title('Comparative Impact on Innovation: GDP per Capita vs. Happiness Score (Simple Regression)', fontsize=14)
plt.xlabel('Coefficient Value (Impact on Innovation Score)', fontsize=12)
plt.ylabel('Factor', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xlim(0, max(coef_gdp, coef_happiness) * 1.1) # Ajusta o limite X
plt.show()

print(f"Coeficiente GDP per Capita: {coef_gdp:.4f}")
print(f"Coeficiente Happiness Score: {coef_happiness:.4f}")
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
import pandas as pd

# --- 1. Padronizar Variveis (Z-Score) ---
Y_innovation_scaled = scale(final_df_clean['Innovation_Score'])
X_gdp_scaled = scale(final_df_clean['GDP_per_capita'])
X_happiness_scaled = scale(final_df_clean['Happiness_Score'])

# --- 2. Ajustar Modelos de Regresso Simples (Padronizados) ---

# Modelo Simples: Inovao vs. GDP_per_capita (Scaled)
X_gdp_const_scaled = sm.add_constant(X_gdp_scaled, has_constant='add')
model_gdp_innovation_scaled = sm.OLS(Y_innovation_scaled, X_gdp_const_scaled).fit()
# O coeficiente do GDP neste modelo  o Coeficiente Beta
coef_gdp_beta = model_gdp_innovation_scaled.params[1] 

# Modelo Simples: Inovao vs. Happiness_Score (Scaled)
X_happiness_const_scaled = sm.add_constant(X_happiness_scaled, has_constant='add')
model_happiness_innovation_scaled = sm.OLS(Y_innovation_scaled, X_happiness_const_scaled).fit()
# O coeficiente da Felicidade neste modelo  o Coeficiente Beta
coef_happiness_beta = model_happiness_innovation_scaled.params[1]

# --- 3. Preparar Dados para Plotagem ---
df_impact_beta = pd.DataFrame({
    'Factor': ['GDP per Capita', 'Happiness Score'],
    'Impact Coefficient (Beta)': [coef_gdp_beta, coef_happiness_beta]
})

# --- 4. Criar Grfico de Barras de Impacto (Beta) ---
plt.figure(figsize=(9, 6))
sns.barplot(
    x='Impact Coefficient (Beta)', 
    y='Factor', 
    data=df_impact_beta, 
    palette=['green', 'red'], 
    orient='h' 
)
plt.title('Comparative Impact on Innovation: GDP vs. Happiness (Standardized Beta Coef.)', fontsize=14)
plt.xlabel('Standardized Impact (Beta Coefficient)', fontsize=12)
plt.ylabel('Factor', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xlim(0, max(coef_gdp_beta, coef_happiness_beta) * 1.1) 
plt.show()

print(f"Coeficiente Beta GDP per Capita: {coef_gdp_beta:.4f}")
print(f"Coeficiente Beta Happiness Score: {coef_happiness_beta:.4f}")
```
